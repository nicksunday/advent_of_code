```python
def create_word_search(file_name: str = 'example1.input.txt') -> list:
    word_search = []
    with open(file_name, 'r', encoding='utf-8') as fin:
        for line in [l for l in fin if len(l.strip()) > 0]:
            word_search.append([c for c in line.strip()])
    return word_search
```

```python
def find_words(word_search) -> int:
    count = 0
    
    for i in range(len(word_search)):
        for j in range(len(word_search[i]) - 3):
            # Forward Horizontal
            if ''.join(word_search[i][j:j+4]) == 'XMAS':
                count += 1
            # Backward Horizontal
            elif ''.join(word_search[i][j:j+4]) == 'SAMX':
                count += 1
    
    horizontal_len = len(word_search[0])
    # Down Vertical
    for i in range(horizontal_len):
        for j in range(len(word_search) - 3):
            word = str(word_search[j][i] + word_search[j+1][i] + word_search[j+2][i] + word_search[j+3][i])
            if word == 'XMAS':
                count += 1
            elif word == 'SAMX':
                count += 1
    
    for i in range(len(word_search) - 3):
        for j in range(len(word_search[i]) - 3):
            # print(f"({i},{j}), ({i + 1},{j + 1}), ({i + 2},{j + 2}), ({i + 3},{j + 3})")
            word = str(word_search[i][j] + word_search[i+1][j+1] + word_search[i+2][j+2] + word_search[i+3][j+3])
            # print(word)
            if word == 'XMAS':
                count += 1
            elif word == 'SAMX':
                count += 1
    
    for i in range(len(word_search) - 1, 2, -1):
        # for j in range(len(word_search[i]) - 1, 2, -1):
        for j in range(0, len(word_search[i]) - 3):
            #print(f"({i},{j}), ({i - 1},{j + 1}), ({i - 2},{j + 2}), ({i - 3},{j + 3})")
            word = str(word_search[i][j] + word_search[i - 1][j + 1] + word_search[i - 2][j + 2] + word_search[i - 3][j + 3])
            # print(f"({i},{j}), ({i + 1},{j + 1}), ({i + 2},{j + 2}), ({i + 3},{j + 3})")
            # word = str(word_search[i][j] + word_search[i+1][j+1] + word_search[i+2][j+2] + word_search[i+3][j+3])
            #print(word)
            if word == 'XMAS':
                count += 1
            elif word == 'SAMX':
                count += 1
    
    return count
```

```python
def print_word_search(word_search) -> None:
    print(" ", [str(n) for n in range(len(word_search[0]))])
    for idx, line in enumerate(word_search):
        print(idx, line)
```

```python
# Example
word_search = create_word_search()
print_word_search(word_search)
print(find_words(word_search))
```
