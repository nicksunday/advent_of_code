```python
pattern="mul\\(\\d{1,3},\\d{1,3}\\)"

# with open("example.input.txt") as fin:
with open("input.txt") as fin:
    muls = re.findall(pattern, fin.read())
    for m in muls: print(m)
```

```python
sum = 0
for m in muls:
    f, s = m.split(",")
    n = re.search("\\d{1,3}", f).group()
    m = re.search("\\d{1,3}", s).group()
    sum += int(n) * int(m)
print(sum)

```
